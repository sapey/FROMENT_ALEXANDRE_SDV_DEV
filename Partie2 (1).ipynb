```python
import pandas as pd
import numpy as np

file = 'mower_market_snapshot.csv'

df = pd.read_csv(file, sep=";")

df
```

```python
def missingValues(dataFrame) :
    print(dataFrame.isna().sum())
    
missingValues(df)
# Fonction pour les valeurs manquantes
```

```python
df1 = df.copy()

def categPrice(categories, dataframe, labels = None):
    dataframe["categ"] = pd.cut(dataframe["price"], categories, labels=labels)
    return dataframe.loc[:,("price", "categ")]
categPrice([0,300, 501, df1.price.max()+1], df1, ["0-200", "201-400", "401++"])
# 3 catégorie de prix
```

```python
df2 = df.copy()

def colonne_binnaire(df, columns, column_to_test, place = 8):
    for col in columns:
        df.insert(place, col, 0)   
    for col in df.loc[:,(columns)]:
        df[col] = df[column_to_test].apply(lambda x : x == col).astype(int)
    return df
        
test =  colonne_binnaire(df2, ["auto-portee", "essence", "electrique"], "product_type")
test
#Colonne binnaire pour les product_type
```

```python
data = dict({"auto-portee" : 1, "essence" : 2, "electrique" : 3})

def numeric(data, df, column):
    for item in data:
        df[column] = df[column].replace(item, data[item])
    return df
d = numeric(data, df2, "product_type")

d
#Transformer donnée en numérique
```
