```python
import pandas as pd
import numpy as numpy
missing_values = ["n/a", "na", "--", "nan", "unknown", " "]
df = pd.read_csv('mower_market_snapshot.csv', sep = ';',na_values = missing_values)
df
```

```python
print(df['prod_cost'].describe())
print(df['prod_cost'].shape)
# Auditer et analyser la colonne prod_cost
```

```python
for x in df:
    print(x ,df[df[x].isnull()].shape[0])
# Nombre de valeur manquante par colonne
```

```python
moyenne = df["prod_cost"].mean()
df["prod_cost"] = df["prod_cost"].fillna(moyenne)
# Remplacer les valeurs manquantes par la moyenne de la colonne
```

```python
df["warranty"].unique()
# Afficher les valeurs uniques de la colonne Warranty
```

```python
df["warranty"].shape
df["warranty"] = df["warranty"].str[0].astype(int)
df["warranty"].unique()
# Traiter la colonne warranty afin d'obtenir une colonne numérique avec 3 valeurs uniques
```

```python
cat1 = df[(df["price"]>0) & (df["price"]<=200)]
cat2 = df[(df["price"]>200) & (df["price"]<=400)]
cat3 = df[df["price"]>400]
# Traiter la colonne price afin d'avoir 3 catégories
```
